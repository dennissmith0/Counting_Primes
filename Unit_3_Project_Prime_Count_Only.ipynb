# PREDICTING PRIME NUMBERS

Prime numbers

(Euclid) A prime number is that which is measured by a unit alone.

A prime number is a number with only two factors, itself and one.

A prime number is only divisible by itself and one.

A prime number cannot be produced by any two numbers (not including itself).

“It is evident that the primes are randomly distributed but, unfortunately, we don’t know what ‘random’ means” (Vaughan, 1990)

```python
import time
import pandas as pd
import numpy as np

start_time = time.time()
num = 10000000
```

```python
# first, lets make a list of all multiples of 6 in our DataFrame range
multiples_6 = [n for n in range(1,num) if n % 6 == 0]
print(multiples_6)
```

```python
#Left hand side column 
  # * NOTE * this gives us ALL LHS values, most of which are NOT prime

LHS = []
for val in multiples_6:
  LHS.append(val - 1)

print(LHS)
```

```python
# RHS column

RHS = []
for val in multiples_6:
  RHS.append(val + 1)
  #print(RHS)

print(RHS)
```
